```python
import pandas as pd
import plotly.express as px
```

```python
df_apps = pd.read_csv("/content/apps.csv")
df_apps.head()
```

```python
df_apps.shape
```

```python
df_apps.drop(['Last_Updated','Android_Ver'],inplace=True)
```

```python
nan_rows = df_apps[df_apps.Rating.isna()]
print(nan_rows.shape)
nan_rows.head()
```

```python
df_apps_clean = df_apps.dropna()
df_apps_clean.shape
```

```python
duplicate_row = df_apps_clean[df_apps_clean.duplicated()]
print(duplicate_row.shape)
duplicate_row.head()
```

```python
df_apps_clean[df_apps_clean.App=="Instagram"]
```

```python
df_apps_clean = df_apps_clean.drop_duplicates()
df_apps_clean[df_apps_clean.App=='Instagram']
```

```python
df_apps_clean = df_apps_clean.drop_duplicates(subset=['App','Type','Price'])
df_apps_clean[df_apps_clean.App=='Instagram']
```

df_apps_clean.sort_values('Rating')
df_apps_clean.head()

```python
df_apps_clean.sort_values('Rating',ascending=False)
df_apps_clean.head()
```

```python
df_apps_clean.sort_values('Size_MBs',ascending=False).head()
```

```python
df_apps_clean.sort_values('Reviews',ascending=False).head()
```

df_apps_clean.sort_values('Price',ascending=True).head()

```python
rating = df_apps_clean.Content_Rating.value_counts()
rating
```

```python
fig = px.pie(labels=rating.index,values=rating.values,title="Content rating ",names=rating.index,hole=0.6)
fig.update_traces(textposition='outside',textinfo='percent+label')
fig.show()
```

```python
df_apps_clean.Installs.describe()
```

```python
df_apps_clean.info()
```

```python
df_apps_clean[['App','Installs']].groupby('Installs').count()
```

```python
df_apps_clean.Installs = df_apps_clean.Installs.astype(str).str.replace(',',"")
df_apps_clean.Installs = pd.to_numeric(df_apps_clean.Installs)
df_apps_clean[['App','Installs']].groupby('Installs').count()
```

```python
df_apps_clean.Price = df_apps_clean.Price.astype(str).str.replace('$',"")
df_apps_clean.Price =pd.to_numeric(df_apps_clean.Price)
df_apps_clean.sort_values('Price',ascending=False).head(20)
```

```python
df_apps_clean.sort_values('Price',ascending=True).tail()
```

```python
df_apps_clean = df_apps_clean[df_apps_clean['Price']<250]
df_apps_clean.sort_values('Price',ascending=False).head(5)
```

```python
df_apps_clean['Revenue_Estimate'] = df_apps_clean.Installs.mul(df_apps_clean.Price)
df_apps_clean.sort_values('Revenue_Estimate', ascending=False)[:10]
```

```python
df_apps_clean.Category.nunique()
```

```python
top_10_category = df_apps_clean.Category.value_counts()[:10]
top_10_category
```

```python
bar = px.bar(x=top_10_category.index,y=top_10_category.values)
bar
```

```python
category_install = df_apps_clean.groupby('Category').agg({'Installs':pd.Series.sum})
category_install.sort_values('Installs',ascending=True,inplace=True)
```

```python
h_bar = px.bar(x=category_install.Installs,y=category_install.index,orientation='h',title='Category Popularity')
h_bar.update_layout(xaxis_title='Number of Dowanloads',yaxis_title='Categroy')

h_bar.show()
```

```python
cat_number = df_apps_clean.groupby('Category').agg({'App': pd.Series.count})
cat_merged_df = pd.merge(cat_number, category_install, on='Category', how="inner")
print(f'The dimensions of the DataFrame are: {cat_merged_df.shape}')
cat_merged_df.sort_values('Installs', ascending=False)
```

scatter = px.scatter(cat_merged_df, # data
                    x='App', # column name
                    y='Installs',
                    title='Category Concentration',
                    size='App',
                    hover_name=cat_merged_df.index,
                    color='Installs')

scatter.update_layout(xaxis_title="Number of Apps (Lower=More Concentrated)",
                      yaxis_title="Installs",
                      yaxis=dict(type='log'))

scatter.show()

```python
stack = df_apps_clean.Genres.str.split(';', expand=True).stack()
print(f'We now have a single column with shape: {stack.shape}')
num_genres = stack.value_counts()
print(f'Number of genres: {len(num_genres)}')
```

```python
bar = px.bar(x = num_genres.index[:15], # index = category name
             y = num_genres.values[:15], # count
             title='Top Genres',
             hover_name=num_genres.index[:15],
             color=num_genres.values[:15],
             color_continuous_scale='Agsunset')

bar.update_layout(xaxis_title='Genre',
yaxis_title='Number of Apps',
coloraxis_showscale=False)

bar.show()
```

```python
df_apps_clean.Type.value_counts()
```

```python
df_free_vs_paid = df_apps_clean.groupby(["Category", "Type"], as_index=False).agg({'App': pd.Series.count})
df_free_vs_paid.head()
```

```python
g_bar = px.bar(df_free_vs_paid,x='Category',y='App',title='Free vs paid by category',color='Type',barmode='group')
g_bar.update_layout(xaxis_title='Category',yaxis_title='Number of Apps',xaxis={'categoryorder':'total descending'},yaxis=dict(type='log'))
g_bar.show()
```

```python
box = px.box(df_apps_clean,
             y='Installs',
             x='Type',
             color='Type',
             notched=True,
             points='all',
             title='How Many Downloads are Paid Apps Giving Up?')

box.update_layout(yaxis=dict(type='log'))

box.show()
```

```python
df_paid_apps = df_apps_clean[df_apps_clean['Type'] == 'Paid']
box = px.box(df_paid_apps, 
             x='Category', 
             y='Revenue_Estimate',
             title='How Much Can Paid Apps Earn?')

box.update_layout(xaxis_title='Category',
                  yaxis_title='Paid App Ballpark Revenue',
                  xaxis={'categoryorder':'min ascending'},
                  yaxis=dict(type='log'))


box.show()
```

```python
box = px.box(df_paid_apps,
             x='Category',
             y="Price",
             title='Price per Category')

box.update_layout(xaxis_title='Category',
                  yaxis_title='Paid App Price',
                  xaxis={'categoryorder':'max descending'},
                  yaxis=dict(type='log'))

box.show()
```
